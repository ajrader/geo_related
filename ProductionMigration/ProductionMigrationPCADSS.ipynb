```python
period = '1011'

def pif_query(period):
    year_1 = int(period[0:2])
    year_2 = int(period[2:4])
    minus_1 = year_1 - 1
    minus_2 = year_1 - 2

    QUERY = """
    select	
        FIPS_CD,
            double( sum( case when year_month > '20$MINUS_1$10'
             and year_month <= '20$YEAR_2$10' then 
             U92.pif_count_new_u(date('20$YEAR_2$-10-31'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) -
             double( sum( case when year_month > '20$MINUS_2$10'
             and year_month <= '20$YEAR_1$10' then 
             U92.pif_count_new_u(date('20$YEAR_1$-10-31'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) 
             as PIF_NET_CHANGE,
        double( sum( case when year_month > '20$MINUS_2$10'
             and year_month <= '20$YEAR_1$10' then 
             U92.pif_count_new_u(date('20$YEAR_1$-10-31'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_1$10,
         double( sum( case when year_month > '20$MINUS_2$11'
             and year_month <= '20$YEAR_1$11' then 
             U92.pif_count_new_u(date('20$YEAR_1$-11-30'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_1$11,
         double( sum( case when year_month > '20$MINUS_2$12'
             and year_month <= '20$YEAR_1$12' then 
             U92.pif_count_new_u(date('20$YEAR_1$-12-31'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_1$12,
         double( sum( case when year_month > '20$MINUS_1$01'
             and year_month <= '20$YEAR_2$01' then 
             U92.pif_count_new_u(date('20$YEAR_2$-01-31'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_2$01,
         double( sum( case when year_month > '20$MINUS_1$02'
             and year_month <= '20$YEAR_2$02' then 
             U92.pif_count_new_u(date('20$YEAR_2$-02-28'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_2$02,
         double( sum( case when year_month > '20$MINUS_1$03'
             and year_month <= '20$YEAR_2$03' then 
             U92.pif_count_new_u(date('20$YEAR_2$-03-31'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_2$03,
         double( sum( case when year_month > '20$MINUS_1$04'
             and year_month <= '20$YEAR_2$04' then 
             U92.pif_count_new_u(date('20$YEAR_2$-04-30'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_2$04,
         double( sum( case when year_month > '20$MINUS_1$05'
             and year_month <= '20$YEAR_2$05' then 
             U92.pif_count_new_u(date('20$YEAR_2$-05-31'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_2$05,
         double( sum( case when year_month > '20$MINUS_1$06'
             and year_month <= '20$YEAR_2$06' then 
             U92.pif_count_new_u(date('20$YEAR_2$-06-30'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_2$06,
         double( sum( case when year_month > '20$MINUS_1$07'
             and year_month <= '20$YEAR_2$07' then 
             U92.pif_count_new_u(date('20$YEAR_2$-07-31'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_2$07,
         double( sum( case when year_month > '20$MINUS_1$08'
             and year_month <= '20$YEAR_2$08' then 
             U92.pif_count_new_u(date('20$YEAR_2$-08-31'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_2$08,
         double( sum( case when year_month > '20$MINUS_1$09'
             and year_month <= '20$YEAR_2$09' then 
             U92.pif_count_new_u(date('20$YEAR_2$-09-30'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_2$09,
         double( sum( case when year_month > '20$MINUS_1$10'
             and year_month <= '20$YEAR_2$10' then 
             U92.pif_count_new_u(date('20$YEAR_2$-10-31'),
             date_effective_date,
             date_expiration_date,
             date_accounting_date,
             transaction_policy_code,
             risk_count) end)) as PIF20$YEAR_2$10	
    from	v23.auto_prem_v as auto_prem_v
    join v23.fips_v as fips_v on 
        auto_prem_v.postl_st_cd = fips_v.postl_st_cd 
        and auto_prem_v.county = fips_v.county_code  
    where	stat_code in ('AA','AC','AM','AN','AP','AR','IA','IC','IM','IN','IP','IR','BA','BC','BM','BN','BP','BR','DA','DC','DM','DN','DP','DR','EA','EC','EM','EN','EP','ER','GA','GC','GM','GN','GP','GR') 
        and year_month between '20$MINUS_2$10' and '20$YEAR_2$10'
        and year between 20$MINUS_2$ and 20$YEAR_2$
        and state not in ('00')
        and vehicle_type in ('0')
    group by FIPS_CD
    order by FIPS_CD
    for fetch only with ur
    """

    query = QUERY.replace('$MINUS_2$', '%02d' % minus_2).replace('$MINUS_1$', '%02d' % minus_1).replace('$YEAR_1$', '%02d' % year_1).replace('$YEAR_2$', '%02d' % year_2)
    return query
```

```python
import ibm_db_dbi as DB2
import sys
import pandas as pd

def pif_df(period):

    DATASOURCE = 'ACTDM4'

    print 'Connecting to %s' % DATASOURCE
    sys.stdout.flush()

    # Open a connection
    conn = DB2.connect(DATASOURCE)

    # Get a cursor, which allows buffered fetching of rows
    curs = conn.cursor()

    print '\tConnected, querying'
    sys.stdout.flush()

    curs.execute(pif_query(period))

    columns = []
    for col in curs.description:
        columns.append(col[0])

    df = pd.DataFrame(columns=columns, data=curs.fetchall())
    df.set_index('FIPS_CD',inplace=True)

    curs.close()

    conn.close()
    
    print '\tDone, %d rows fetched' % len(df)
    
    return df
```
