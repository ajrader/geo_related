```python
#read in the netflow info
def getFLOW(pth,fname,period,fmt='.csv'):
    
    df = pd.read_csv(pth+fname+period+fmt,usecols=range(1,16),
                      dtype={'State_Code':np.str, 'County_Code':np.str,
                             'Code1':np.str, 'Code2':np.str})
    # assign FIPS_CD
    df['FIPS_CD'] = df.State_Code + df.County_Code
    # assign ITEM_CD (i.e. migrants or non-migrants, etc.) --> 
    # for first pass probably just want non_migrants (ITEM_CD == FIPS_CD) and total_migrants (96000)
    df['ITEM_CD'] = df.Code1 + df.Code2
    df['FIPS']=df['FIPS_CD']
    df['Period']=period
    fdf = df.drop(['State_Code','County_Code','Code1','Code2'], axis=1).set_index(['ITEM_CD', 'FIPS_CD'])
    #fdf = df.drop(['State_Code','County_Code','Code1','Code2'], axis=1)
    #assign Time Period to this dataframe
    
    return fdf
```

```python
# function to combine the results from different years for each fipscode
def createComboResults(fdf):
    # group into total_migrants and non-Migrants
    groups_fdf = full_fdf.groupby(by=
                         lambda x: '99999' if x[0]==x[1] else x[0] )

    total_migrants = groups_fdf.get_group('96000')
    #total_migrants.drop('index',1,inplace=True)
    non_migrants = groups_fdf.get_group('99999')#.reset_index(level=0)#.reset_index(level='ITEM_CD', drop=True)
    # get a list of unique FIPS codes for these set
    fipsList_nm = non_migrants.FIPS.unique()
    fipsList_tm = total_migrants.FIPS.unique()
    print len(fipsList_tm), len(fipsList_nm)
    fipsIntersect = set(fipsList_tm).intersection(set(fipsList_nm))
    cols2sum = ['Return_Out','Return_In','Return_Net','Exmpt_Out','Exmpt_In','Exmpt_Net','Aggr_AGI_Out','Aggr_AGI_In','Aggr_AGI_Net']
    
    #cols2pull = ['Period','FIPS','Abbrv']
    mcols = [a for a in cols2sum]
    mcols.append('FIPS')
    mcols.append('Period')
    mcols.append('Abbrv')
    jDF = pd.DataFrame(columns=mcols)
    for f in fipsIntersect:
        #joinedDF=pd.DataFrame(columns=[cols2pull+sumColsRename])
        # TODO check that  the number of elements is the same in both
        #print f, sum(non_migrants['FIPS']==f),sum(total_migrants['FIPS']==f)
        
        #nm01001b = non_migrants2[non_migrants2['FIPS']=='01001'].copy()
#tm01001b = total_migrants2[total_migrants2['FIPS']=='01001'].copy()
        nm = non_migrants[non_migrants['FIPS']==f]#.copy()
        tm = total_migrants[total_migrants['FIPS']==f]#.copy()
        state = tm['Abbrv'].unique()[0]
        nm.index = nm.Period
        tm.index = tm.Period
        joinedDF= nm[cols2sum]+tm[cols2sum]
        #AA = tm[cols2pull].copy()
        
        joinedDF['FIPS']= f
        joinedDF['Period']=tm.Period
        joinedDF['Abbrv']=state
        
        #print f, joinedDF.ix['0405']
        #jDF.append(joinedDF,ignore_index=True)
        #print f, AA.reset_index(drop=True)
        #joinedDF[cols2pull]= tm[cols2pull].copy()
        #print f, np.shape(joinedDF),np.shape(jDF)
        jDF=pd.concat([jDF,joinedDF])
        
    # rename the columns: create a dictionary
    sumColsRename= ['Return_Start','Return_End','DeltaReturn','Exmpt_Start','Exmpt_End','DeltaExmpt','Aggr_AGI_Start','Aggr_AGI_End','DeltaAggr_AGI']
    cDict = {}
    for ic in xrange(0,len(cols2sum)):
        cDict[cols2sum[ic]]=sumColsRename[ic]
        
    
    jDF.rename(columns=cDict,inplace=True)
    
    return jDF
        
```

```python
## helper plotting function
def countyScatter(df1,df2,countyID,color='steelblue'):
    clineColor = 'lightcoral'
    l1 = len(df1[countyID])
    l2 = len(df2[countyID])
    # assumes that the extra data is on the tail (more recent) end
    s1 = df1[countyID].values
    s2 = df2[countyID].values
    if l2 >l1:
        s2 = df2[countyID].values[:l1]
    elif l1 > l2:
        s1 = df1[countyID].values[:l2]
    plt.scatter(s1,s2,color=color)
    plt.grid(True)
    cc = np.corrcoef(s1,s2)[1][0]
    #print cc
    print "Correlation Coefficient is {0}".format(cc)
    return cc
```

```python
def fips_to_id(fips):
    v = np.int(fips)
    v1 = v / 1000
    v2 = np.mod(v,1000)
    return (v1,v2)

def map_data(df, column, plot_title, file_name, quartile_scale=0.05, create_map=True,raw_scale=False):
    from matplotlib.colors import LogNorm, Normalize, rgb2hex
    from matplotlib.cm import ScalarMappable
    import matplotlib.colors as col
    import bokeh.plotting as bp
    from bokeh.sampledata import us_counties, us_states

    us_states = us_states.data.copy()
    us_counties = us_counties.data.copy()

    del us_states['HI']
    del us_states['AK']

    state_xs = [us_states[code]['lons'] for code in us_states]
    state_ys = [us_states[code]['lats'] for code in us_states]


    df['county_id'] = map(fips_to_id, df.index.values)

    df = df.reset_index().set_index('county_id')


    county_ids=list([code for code in us_counties if us_counties[code]['state'] not in ['ak', 'hi', 'pr', 'gu', 'vi', 'mp', 'as']])

    df = df.ix[county_ids]

    county_xs=list([us_counties[code]['lons'] for code in us_counties if us_counties[code]['state'] not in ['ak', 'hi', 'pr', 'gu', 'vi', 'mp', 'as']])
    county_ys=list([us_counties[code]['lats'] for code in us_counties if us_counties[code]['state'] not in ['ak', 'hi', 'pr', 'gu', 'vi', 'mp', 'as']])

    df.loc[county_ids, 'lons'] = county_xs
    df.loc[county_ids, 'lats'] = county_ys


    clist = ['#fc8d59',
            '#ffffbf',
            '#91bfdb']
    
    under_min = '#fb580c'
    over_max = '#4794c3'

    cmap = col.LinearSegmentedColormap.from_list('colors',clist)
    cmap.set_over(color=over_max)
    cmap.set_under(color=under_min)

    #df = pc_migration.Aggr_AGI
    #df = pc_migration.Return
    df_col = df[column] * 100.0

    orig_min=df_col.quantile(q=quartile_scale)
    orig_max=df_col.quantile(q=1-quartile_scale)
    
    if raw_scale:
        print orig_max,orig_min
        
    else:
     
        if (orig_min<0) and (orig_max>0) and (np.abs(orig_min) < np.abs(orig_max)): 
            orig_min = -orig_max 
        else:
            orig_max = -orig_min
        
        
    norm = Normalize(vmin=orig_min, vmax=orig_max)
    sm = ScalarMappable(norm=norm, cmap=cmap)

    sm.set_array(df_col)
    c = sm.to_rgba(df_col)
    df['color'] = map(rgb2hex,c)
    
    if create_map:
        from matplotlib.colorbar import colorbar_factory
        import bokeh.mpl as mpl

        bp.reset_output()

        bp.output_file(file_name, title=plot_title)

        bp.hold()

        bp.patches(df.lons, df.lats, fill_color=df.color, fill_alpha=0.7,
                line_color="white", line_width=0.5, plot_width=1100, plot_height=700,
                title=plot_title)
        bp.patches(state_xs, state_ys, fill_alpha=0.0, line_color="#884444",
                line_width=2, name="choropleth")

        bp.hold(value=False)
       
        bp.show()
        bp.reset_output()
        
        f = plt.figure()
        rect = 0,0,1,0.1
        ax = f.add_axes(rect)
        #f.colorbar(sm, extend='both', format='%0.2f%%')
        colorbar_factory(ax,sm,extend='both', format='%0.2f%%', orientation='horizontal')
        f.show()

    
   
```

### Normalize everything by state

```python
quartile_scale=0.05

state_pc = pc_migration
state_pc['PIF_PERCENT_CHANGE'] = pif_df_numeric.PIF_PERCENT_CHANGE

state_pc['state'] = map(lambda x: x[0:2], state_pc.index.values)
state_pc.sort_index(inplace=True)

state_pc_min = state_pc.groupby('state').transform(lambda x: x.quantile(quartile_scale))
state_pc_max = state_pc.groupby('state').transform(lambda x: x.quantile(1-quartile_scale))

def thresh(x, x_min, x_max):
    if x.name not in x_min.columns:
        return x
    return np.max([np.min([x.values,x_max[x.name].values], axis=0), x_min[x.name].values],axis=0)
    
state_pc_thresh = state_pc.apply(thresh,args=(state_pc_min, state_pc_max))

def norm(x):
    y = (x - x.mean()) / x.std()
    return y.fillna(value=0.0)

state_pc_norm = state_pc_thresh.groupby('state').transform(norm)
state_pc_norm['state'] = map(lambda x: x[0:2], state_pc_norm.index.values)
```

## Select Features

```python
import sklearn.feature_selection as fs

X = state_pc_norm[['Aggr_AGI','Return','Exmpt']]
Y = state_pc_norm.PIF_PERCENT_CHANGE

def rate_features(X,Y, plot=False):
    selector = fs.SelectKBest(score_func=fs.f_regression, k='all')
    X_indices = np.arange(X.shape[-1])

    selector.fit(X,Y)
    scores = -np.log10(selector.pvalues_)
    scores /= scores.max()
    
    if plot:
        fig, ax = plt.subplots()
        ax.bar(X_indices - .45, scores, width=.2,
               label=r'Univariate score ($-Log(p_{value})$)', color='g')
        ax.set_xticks(X_indices - .45+.2)
        ax.set_xticklabels(X.columns)
    return scores
```

```python
rate_features(X,Y)
```

## Look at per state importance

```python
def per_state_scores(df):
    sg = df.groupby('state')
    scores = pd.DataFrame(index=sg.groups.keys(), columns=['Aggr_AGI','Return','Exmpt'])
    for g in sg.groups.keys():
        group = sg.get_group(g)
        scores.loc[g, ['Aggr_AGI','Return','Exmpt']] = rate_features(group[['Aggr_AGI','Return','Exmpt']], group['PIF_PERCENT_CHANGE'])

    scores = scores.fillna(scores.mean())
    sm = scores.apply(np.mean)
    sv = scores.apply(np.std)

    fig, ax = plt.subplots()
    ax.bar([-1,0,1], sm, width=.2,
           label=r'Univariate score ($-Log(p_{value})$)', color='grey', yerr=sv)
    ax.set_xticks([-1,0,1])
    ax.set_xticklabels(scores.columns)

    return scores
```

## DYNAMIC TIME WARPING ASIDE

```python
import sys
import collections
import itertools
from scipy.stats import mode

try:
    from IPython.display import clear_output
    have_ipython = True
except ImportError:
    have_ipython = False

class KnnDtw(object):
    """K-nearest neighbor classifier using dynamic time warping
    as the distance measure between pairs of time series arrays
    
    Arguments
    ---------
    n_neighbors : int, optional (default = 5)
        Number of neighbors to use by default for KNN
        
    max_warping_window : int, optional (default = infinity)
        Maximum warping window allowed by the DTW dynamic
        programming function
            
    subsample_step : int, optional (default = 1)
        Step size for the timeseries array. By setting subsample_step = 2,
        the timeseries length will be reduced by 50% because every second
        item is skipped. Implemented by x[:, ::subsample_step]
    """
    
    def __init__(self, n_neighbors=5, max_warping_window=10000, subsample_step=1):
        self.n_neighbors = n_neighbors
        self.max_warping_window = max_warping_window
        self.subsample_step = subsample_step
    
    def fit(self, x, l):
        """Fit the model using x as training data and l as class labels
        
        Arguments
        ---------
        x : array of shape [n_samples, n_timepoints]
            Training data set for input into KNN classifer
            
        l : array of shape [n_samples]
            Training labels for input into KNN classifier
        """
        
        self.x = x
        self.l = l
        
    def _dtw_distance(self, ts_a, ts_b, d = lambda x,y: abs(x-y)):
        """Returns the DTW similarity distance between two 2-D
        timeseries numpy arrays.

        Arguments
        ---------
        ts_a, ts_b : array of shape [n_samples, n_timepoints]
            Two arrays containing n_samples of timeseries data
            whose DTW distance between each sample of A and B
            will be compared
        
        d : DistanceMetric object (default = abs(x-y))
            the distance measure used for A_i - B_j in the
            DTW dynamic programming function
        
        Returns
        -------
        DTW distance between A and B
        """

        # Create cost matrix via broadcasting with large int
        ts_a, ts_b = np.array(ts_a), np.array(ts_b)
        M, N = len(ts_a), len(ts_b)
        cost = sys.maxint * np.ones((M, N))

        # Initialize the first row and column
        cost[0, 0] = d(ts_a[0], ts_b[0])
        for i in xrange(1, M):
            cost[i, 0] = cost[i-1, 0] + d(ts_a[i], ts_b[0])

        for j in xrange(1, N):
            cost[0, j] = cost[0, j-1] + d(ts_a[0], ts_b[j])

        # Populate rest of cost matrix within window
        for i in xrange(1, M):
            for j in xrange(max(1, i - self.max_warping_window),
                            min(N, i + self.max_warping_window)):
                choices = cost[i - 1, j - 1], cost[i, j-1], cost[i-1, j]
                cost[i, j] = min(choices) + d(ts_a[i], ts_b[j])

        # Return DTW distance given window 
        return cost[-1, -1]
    
    def _dist_matrix(self, x, y):
        """Computes the M x N distance matrix between the training
        dataset and testing dataset (y) using the DTW distance measure
        
        Arguments
        ---------
        x : array of shape [n_samples, n_timepoints]
        
        y : array of shape [n_samples, n_timepoints]
        
        Returns
        -------
        Distance matrix between each item of x and y with
            shape [training_n_samples, testing_n_samples]
        """
        
        # Compute the distance matrix        
        dm_count = 0
        
        # Compute condensed distance matrix (upper triangle) of pairwise dtw distances
        # when x and y are the same array
        if(np.array_equal(x, y)):
            x_s = np.shape(x)
            dm = np.zeros((x_s[0] * (x_s[0] - 1)) // 2, dtype=np.double)
            
            p = ProgressBar(np.shape(dm)[0])
            
            for i in xrange(0, x_s[0] - 1):
                for j in xrange(i + 1, x_s[0]):
                    dm[dm_count] = self._dtw_distance(x[i, ::self.subsample_step],
                                                      y[j, ::self.subsample_step])
                    
                    dm_count += 1
                    p.animate(dm_count)
            
            # Convert to squareform
            dm = squareform(dm)
            return dm
        
        # Compute full distance matrix of dtw distnces between x and y
        else:
            x_s = np.shape(x)
            y_s = np.shape(y)
            dm = np.zeros((x_s[0], y_s[0])) 
            dm_size = x_s[0]*y_s[0]
            
            p = ProgressBar(dm_size)
        
            for i in xrange(0, x_s[0]):
                for j in xrange(0, y_s[0]):
                    dm[i, j] = self._dtw_distance(x[i, ::self.subsample_step],
                                                  y[j, ::self.subsample_step])
                    # Update progress bar
                    dm_count += 1
                    p.animate(dm_count)
        
            return dm
        
    def predict(self, x):
        """Predict the class labels or probability estimates for 
        the provided data

        Arguments
        ---------
          x : array of shape [n_samples, n_timepoints]
              Array containing the testing data set to be classified
          
        Returns
        -------
          2 arrays representing:
              (1) the predicted class labels 
              (2) the knn label count probability
        """
        
        dm = self._dist_matrix(x, self.x)

        # Identify the k nearest neighbors
        knn_idx = dm.argsort()[:, :self.n_neighbors]

        # Identify k nearest labels
        knn_labels = self.l[knn_idx]
        
        # Model Label
        mode_data = mode(knn_labels, axis=1)
        mode_label = mode_data[0]
        mode_proba = mode_data[1]/self.n_neighbors

        return mode_label.ravel(), mode_proba.ravel()

class ProgressBar:
    """This progress bar was taken from PYMC
    """
    def __init__(self, iterations):
        self.iterations = iterations
        self.prog_bar = '[]'
        self.fill_char = '*'
        self.width = 40
        self.__update_amount(0)
        if have_ipython:
            self.animate = self.animate_ipython
        else:
            self.animate = self.animate_noipython

    def animate_ipython(self, iter):
        print '\r', self,
        sys.stdout.flush()
        self.update_iteration(iter + 1)

    def update_iteration(self, elapsed_iter):
        self.__update_amount((elapsed_iter / float(self.iterations)) * 100.0)
        self.prog_bar += '  %d of %s complete' % (elapsed_iter, self.iterations)

    def __update_amount(self, new_amount):
        percent_done = int(round((new_amount / 100.0) * 100.0))
        all_full = self.width - 2
        num_hashes = int(round((percent_done / 100.0) * all_full))
        self.prog_bar = '[' + self.fill_char * num_hashes + ' ' * (all_full - num_hashes) + ']'
        pct_place = (len(self.prog_bar) // 2) - len(str(percent_done))
        pct_string = '%d%%' % percent_done
        self.prog_bar = self.prog_bar[0:pct_place] + \
            (pct_string + self.prog_bar[pct_place + len(pct_string):])

    def __str__(self):
        return str(self.prog_bar)
```
